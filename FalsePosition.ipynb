```python
def fun(j):
  f=(j**6)-j-1;
  return f

def false_position_method(t):
    i = 0
    j = 0
    while fun(j) * fun(j + 1) > 0:
        j += 1
    a = j
    b = j + 1
    print("a : ", a)
    print("b : ", b)
    c = (a * fun(b) - b * fun(a)) / (fun(b) - fun(a))
    previous_c = None
    while True:
        previous_c = c
        f = fun(c)
        if f < 0:
            a = c
        elif f > 0:
            b = c
        c = (a * fun(b) - b * fun(a)) / (fun(b) - fun(a))
        i += 1
        print("Iteration : ", i)
        print("a : ", a)
        print("b : ", b)
        print("c : ", c)
        if abs(previous_c - c) <= t:
            break
    print("root : ", c)

false_position_method(0.00005)
```
